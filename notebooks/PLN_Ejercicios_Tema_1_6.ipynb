### Aplicación 1: Diagnóstico Clínico Asistido por PLN en Historiales Médicos

**¿En qué consiste?**  
Sistemas como **Amazon Comprehend Medical** o **Google Cloud Healthcare NLP API** analizan automáticamente historiales clínicos en texto libre (notas de médicos, informes de alta, resultados de laboratorio) para extraer entidades médicas: diagnósticos, medicamentos, dosis, síntomas y relaciones entre ellos.

**¿Por qué es útil e interesante?**  
Los médicos documentan en lenguaje natural, no en bases de datos estructuradas. El PLN permite convertir ese texto no estructurado en información accionable: detectar interacciones medicamentosas, identificar pacientes de riesgo, acelerar la codificación ICD-10 para facturación, y apoyar la investigación epidemiológica. Es una aplicación donde el PLN salva vidas literalmente, al reducir errores médicos por información dispersa o mal codificada.

## Profundización en un Caso: Generación Automática de Código

**Caso elegido:** GitHub Copilot — desarrollado por GitHub en colaboración con OpenAI.

### Descripción del caso concreto

**GitHub Copilot** fue lanzado en 2021 y se convirtió en el asistente de código con mayor adopción en la historia del desarrollo de software. En 2023, GitHub reportó que más de **1 millón de desarrolladores** lo usaban activamente, y que el 46% del código en los proyectos donde estaba habilitado era generado por la IA.

### ¿Cómo funciona la parte de PLN?

GitHub Copilot se basa en **OpenAI Codex**, un modelo de la familia GPT entrenado específicamente sobre código fuente. La arquitectura es un **Transformer** de tipo decoder-only (similar a GPT-3/4), con las siguientes características clave:

1. **Preentrenamiento masivo:** El modelo fue entrenado sobre decenas de millones de repositorios públicos de GitHub, aprendiendo patrones de código en más de 12 lenguajes de programación. Esto le da una comprensión profunda de la sintaxis, semántica y convenciones de cada lenguaje.

2. **Comprensión del contexto:** El modelo recibe como entrada el código ya escrito en el archivo (contexto), los comentarios en lenguaje natural del desarrollador, y el nombre de la función o variable que se está escribiendo. Con esa información, predice los tokens más probables que completan el código.

3. **Generación autoregresiva:** El código se genera token a token (palabra por palabra, símbolo por símbolo), donde cada token generado se añade al contexto para predecir el siguiente. Esto es exactamente el mismo mecanismo que usa GPT para generar texto.

4. **Fine-tuning con RLHF:** El modelo fue refinado con feedback humano (Reinforcement Learning from Human Feedback), donde programadores evaluaron la calidad de las sugerencias, mejorando la relevancia y corrección del código generado.

## Paso 1 — Importaciones y descarga de recursos NLTK

Importamos todas las librerías de una vez (buena práctica PEP8). Descargamos los recursos de NLTK que necesitamos: `punkt` para tokenización y `stopwords` para el filtrado.

```python
import re
import string
import time

import nltk
import spacy
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer

# Descarga de recursos NLTK necesarios
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)

# Carga del modelo de spaCy en español
# es_core_news_sm: modelo pequeño entrenado sobre noticias en español
nlp = spacy.load('es_core_news_sm')

print('✓ Librerías cargadas correctamente')
```

## Paso 2 — Corpus de texto

Definimos un corpus de 5 reseñas de películas en español. Se eligió este dominio porque:
- Contiene lenguaje informal y formal mezclado
- Tiene signos de puntuación variados (¡!, ¿?)
- Incluye palabras con mayúsculas expresivas
- Permite ver claramente el efecto de cada técnica de preprocesamiento

```python
# Corpus: 5 reseñas de películas en español
# Variedad intencional: longitud, tono, puntuación y vocabulario distintos
corpus = [
    "¡Esta película es INCREÍBLE! Los actores actuaron de manera extraordinaria y la dirección fue magistral.",
    "Muy aburrida y predecible. No recomendaría esta película a nadie. Perdí 2 horas de mi vida.",
    "Los efectos especiales son impresionantes, pero la historia deja mucho que desear. Actuaciones mediocres.",
    "Una obra maestra del cine español. Emotiva, profunda y con un guión perfectamente construido.",
    "Regular tirando a mala. El director claramente no sabía lo que hacía. Diálogos forzados y sin sentido."
]

print(f'Corpus cargado: {len(corpus)} documentos')
print('\n--- Textos originales ---')
for i, texto in enumerate(corpus, 1):
    print(f'[{i}] {texto}')
```

## Paso 3 — Normalización

**¿Qué es?** Convertir el texto a una forma estándar: minúsculas, sin caracteres especiales, sin números.

**¿Por qué primero?** La normalización debe hacerse antes de tokenizar para que 'INCREÍBLE' e 'increíble' sean tratadas como la misma palabra. Si tokenizamos primero, tendríamos tokens duplicados.

**Decisión de diseño:** Mantenemos los acentos (á, é, í, ó, ú) porque en español son semánticamente relevantes ('mas' ≠ 'más'). Eliminamos signos de puntuación especiales del español (¡¿) pero mantenemos la estructura de las palabras.

```python
def normalizar(texto):
    """
    Normaliza un texto aplicando:
    1. Conversión a minúsculas
    2. Eliminación de signos de puntuación (incluidos ¡¿)
    3. Eliminación de números
    4. Eliminación de espacios múltiples
    """
    texto = texto.lower()                          # minúsculas
    texto = re.sub(r'[¡¿!?.,;:\"\'-]', ' ', texto) # eliminar puntuación
    texto = re.sub(r'\d+', '', texto)              # eliminar números
    texto = re.sub(r'\s+', ' ', texto).strip()     # espacios múltiples → uno
    return texto

corpus_normalizado = [normalizar(t) for t in corpus]

print('--- Textos normalizados ---')
for i, (orig, norm) in enumerate(zip(corpus, corpus_normalizado), 1):
    print(f'[{i}] ORIGINAL: {orig}')
    print(f'    NORMALIZADO: {norm}')
    print()
```

## Paso 4 — Tokenización

**¿Qué es?** Dividir el texto en unidades mínimas (tokens), que generalmente son palabras.

**¿Por qué `word_tokenize` de NLTK?** Es más sofisticado que un simple `split()`: maneja contracciones, abreviaciones y puntuación de forma inteligente. Usa el modelo Punkt entrenado para el idioma.

**Comparación:** También mostramos tokenización por oraciones (`sent_tokenize`) para ilustrar que la tokenización puede operar a distintos niveles.

```python
# Tokenización a nivel de palabra sobre el corpus normalizado
corpus_tokenizado = [word_tokenize(t, language='spanish') for t in corpus_normalizado]

print('--- Tokenización por palabras ---')
for i, tokens in enumerate(corpus_tokenizado, 1):
    print(f'[{i}] {tokens}')
    print(f'    → {len(tokens)} tokens')
    print()

# Tokenización por oraciones (sobre texto original, antes de normalizar)
print('--- Tokenización por oraciones (texto original) ---')
for i, texto in enumerate(corpus, 1):
    oraciones = sent_tokenize(texto, language='spanish')
    print(f'[{i}] {oraciones}')
```

## Paso 5 — Eliminación de Stopwords

**¿Qué son las stopwords?** Palabras muy frecuentes que aportan poco significado semántico: 'de', 'la', 'que', 'y', 'a', 'en'...

**¿Por qué eliminarlas?** Reducen el ruido en modelos de clasificación y representación. Si no las eliminamos, 'de' y 'la' serían las palabras más frecuentes en cualquier corpus español, dominando las representaciones numéricas sin aportar información discriminativa.

**Decisión de diseño:** Usamos la lista de stopwords de NLTK para español. En un proyecto real, podríamos añadir stopwords específicas del dominio (ej: 'película', 'film' si solo analizamos reseñas de cine).

```python
# Carga de stopwords en español de NLTK
stop_words = set(stopwords.words('spanish'))
print(f'Total stopwords en español (NLTK): {len(stop_words)}')
print(f'Ejemplos: {list(stop_words)[:15]}')
print()

def eliminar_stopwords(tokens):
    """Filtra tokens que sean stopwords o tengan longitud <= 1."""
    return [t for t in tokens if t not in stop_words and len(t) > 1]

corpus_sin_stop = [eliminar_stopwords(tokens) for tokens in corpus_tokenizado]

print('--- Tokens sin stopwords ---')
for i, (con, sin) in enumerate(zip(corpus_tokenizado, corpus_sin_stop), 1):
    eliminadas = len(con) - len(sin)
    porcentaje = (eliminadas / len(con) * 100) if len(con) > 0 else 0
    print(f'[{i}] {sin}')
    print(f'    → Eliminados {eliminadas} tokens ({porcentaje:.1f}% del total)')
    print()
```

## Paso 6 — Stemming

**¿Qué es?** Reducir cada palabra a su raíz morfológica mediante reglas heurísticas. No produce palabras reales, sino raíces.

**Algoritmo:** `SnowballStemmer` para español (también llamado Spanish Stemmer). Es una mejora del algoritmo de Porter, adaptado para lenguas romances.

**Ventaja:** Muy rápido, sin necesidad de diccionario.
**Desventaja:** Puede producir raíces que no son palabras reales ('actuaron' → 'actu'), perdiendo legibilidad.

```python
# Inicialización del stemmer para español
stemmer = SnowballStemmer('spanish')

def aplicar_stemming(tokens):
    """Aplica stemming a cada token de la lista."""
    return [stemmer.stem(t) for t in tokens]

corpus_stemmed = [aplicar_stemming(tokens) for tokens in corpus_sin_stop]

print('--- Stemming aplicado ---')
for i, (antes, despues) in enumerate(zip(corpus_sin_stop, corpus_stemmed), 1):
    print(f'[{i}] ANTES:  {antes}')
    print(f'    DESPUÉS: {despues}')
    # Mostrar pares token → stem para análisis
    pares = [(a, d) for a, d in zip(antes, despues) if a != d]
    if pares:
        print(f'    Cambios: {pares}')
    print()
```

## Paso 7 — Lematización

**¿Qué es?** Reducir cada palabra a su forma canónica (lema) usando un diccionario morfológico y análisis gramatical. A diferencia del stemming, el lema siempre es una palabra real.

**Herramienta:** spaCy con el modelo `es_core_news_sm`. Este modelo usa redes neuronales para determinar el lema correcto según el contexto gramatical.

**Ejemplo clave:** 'actuaron' → lema 'actuar' (correcto), mientras que el stemmer daría 'actu' (no es una palabra).

**Nota importante:** La lematización se aplica sobre el texto normalizado (no sobre los tokens ya filtrados) porque spaCy necesita el contexto de la oración para determinar el lema correcto.

```python
def lematizar(texto):
    """
    Lematiza un texto usando spaCy.
    Retorna lista de lemas, excluyendo stopwords y puntuación.
    spaCy necesita el texto completo (no tokens) para analizar el contexto.
    """
    doc = nlp(texto)
    return [
        token.lemma_.lower()
        for token in doc
        if not token.is_stop       # excluir stopwords
        and not token.is_punct     # excluir puntuación
        and not token.is_space     # excluir espacios
        and len(token.text) > 1    # excluir tokens de 1 carácter
    ]

corpus_lematizado = [lematizar(t) for t in corpus_normalizado]

print('--- Lematización aplicada ---')
for i, (norm, lemas) in enumerate(zip(corpus_normalizado, corpus_lematizado), 1):
    print(f'[{i}] NORMALIZADO: {norm}')
    print(f'    LEMAS:       {lemas}')
    print()
```

## Paso 8 — Comparación directa: Stemming vs Lematización

Esta comparación es clave para entender la diferencia fundamental entre ambas técnicas.

```python
# Comparación directa palabra por palabra: stemming vs lematización
# Usamos el primer documento como ejemplo representativo
print('=== COMPARACIÓN: Stemming vs Lematización (Documento 1) ===')
print(f'Original: {corpus[0]}')
print()

tokens_ejemplo = corpus_sin_stop[0]
stems_ejemplo = corpus_stemmed[0]
lemas_ejemplo = corpus_lematizado[0]

# Tabla comparativa
print(f'{"Token original":<20} {"Stem":<20} {"Lema":<20}')
print('-' * 60)
for token, stem in zip(tokens_ejemplo, stems_ejemplo):
    # Buscar el lema correspondiente (aproximación)
    doc = nlp(token)
    lema = doc[0].lemma_.lower() if doc else token
    print(f'{token:<20} {stem:<20} {lema:<20}')

print()
print('OBSERVACIÓN: El stemming produce raíces no siempre legibles,')
print('mientras que la lematización produce formas canónicas reales del diccionario.')
```

## Paso 9 — Pipeline completo y resumen estadístico

```python
# Resumen estadístico del pipeline completo
print('=== RESUMEN ESTADÍSTICO DEL PIPELINE ===')
print(f'{"Etapa":<30} {"Tokens totales":<20} {"Reducción"}')
print('-' * 65)

total_orig = sum(len(word_tokenize(t, language='spanish')) for t in corpus)
total_norm = sum(len(t) for t in corpus_tokenizado)
total_stop = sum(len(t) for t in corpus_sin_stop)
total_stem = sum(len(t) for t in corpus_stemmed)
total_lema = sum(len(t) for t in corpus_lematizado)

etapas = [
    ('1. Texto original', total_orig, 0),
    ('2. Normalizado + tokenizado', total_norm, total_orig - total_norm),
    ('3. Sin stopwords', total_stop, total_orig - total_stop),
    ('4. Stemming', total_stem, total_orig - total_stem),
    ('5. Lematización', total_lema, total_orig - total_lema),
]

for nombre, total, reduccion in etapas:
    pct = (reduccion / total_orig * 100) if total_orig > 0 else 0
    print(f'{nombre:<30} {total:<20} {reduccion:+d} ({pct:.1f}%)')
```

## Análisis y conclusiones

### Resultados observados

**Normalización:** La conversión a minúsculas y eliminación de puntuación redujo el vocabulario efectivo al unificar variantes como 'INCREÍBLE' e 'increíble'. Los signos ¡¿ del español son especialmente importantes de eliminar porque NLTK no siempre los maneja bien en tokenización.

**Tokenización:** `word_tokenize` de NLTK funcionó correctamente para el español, separando correctamente palabras y signos de puntuación. La tokenización por oraciones (`sent_tokenize`) fue menos precisa con oraciones cortas separadas por puntos seguidos de mayúsculas.

**Stemming vs Lematización — diferencia clave:**
- El stemmer convirtió 'actuaron' → 'actu', 'extraordinaria' → 'extraordinari'. Estas raíces no son palabras reales pero son útiles para agrupar variantes morfológicas.
- La lematización convirtió 'actuaron' → 'actuar', 'extraordinaria' → 'extraordinario'. Produce formas reales del diccionario, más interpretables.
- **¿Cuándo usar cada uno?** Stemming: cuando la velocidad es crítica y la interpretabilidad no importa (motores de búsqueda a gran escala). Lematización: cuando necesitamos resultados interpretables o trabajamos con modelos que usan vocabulario real (embeddings preentrenados).

## Paso 0 — Instalación e importaciones

```python
import numpy as np
import pandas as pd
import math
from collections import Counter

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.preprocessing import LabelBinarizer

nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)

print('✓ Librerías cargadas')
```

## Paso 1 — Corpus y preprocesamiento base

Reutilizamos el corpus de reseñas del Tema 2. Aplicamos normalización básica antes de las representaciones.

```python
import re

corpus = [
    "Esta película es increíble los actores actuaron de manera extraordinaria",
    "Muy aburrida y predecible no recomendaría esta película a nadie",
    "Los efectos especiales son impresionantes pero la historia deja mucho que desear",
    "Una obra maestra del cine español emotiva profunda y con guión perfectamente construido",
    "Regular tirando a mala el director no sabía lo que hacía diálogos forzados"
]

stop_es = set(stopwords.words('spanish'))

def preprocesar(texto):
    """Normaliza y elimina stopwords para representaciones de texto."""
    texto = texto.lower()
    texto = re.sub(r'[^a-záéíóúüñ\s]', '', texto)
    tokens = word_tokenize(texto, language='spanish')
    tokens = [t for t in tokens if t not in stop_es and len(t) > 2]
    return ' '.join(tokens)

corpus_prep = [preprocesar(t) for t in corpus]

print('Corpus preprocesado:')
for i, t in enumerate(corpus_prep, 1):
    print(f'[{i}] {t}')
```

## Representación 1 — Bag of Words (BoW)

**¿Qué es?** Representa cada documento como un vector de conteos de palabras. El vocabulario es el conjunto de todas las palabras únicas del corpus. Cada posición del vector corresponde a una palabra, y el valor es cuántas veces aparece en ese documento.

**Ventaja:** Simple, rápido, interpretable.
**Desventaja:** Ignora el orden de las palabras y la frecuencia relativa entre documentos. 'buena' y 'mala' tienen el mismo peso si aparecen el mismo número de veces.

```python
# CountVectorizer implementa Bag of Words eficientemente
# binary=False → conteos reales (no solo presencia/ausencia)
bow_vectorizer = CountVectorizer()
bow_matrix = bow_vectorizer.fit_transform(corpus_prep)

vocabulario = bow_vectorizer.get_feature_names_out()
bow_df = pd.DataFrame(
    bow_matrix.toarray(),
    columns=vocabulario,
    index=[f'Doc{i+1}' for i in range(len(corpus))]
)

print(f'Vocabulario total: {len(vocabulario)} palabras únicas')
print(f'Dimensión de la matriz: {bow_matrix.shape}')
print()
print('Matriz Bag of Words:')
print(bow_df.to_string())
```

## Representación 2 — One-Hot Encoding

**¿Qué es?** Variante de BoW donde solo se registra la presencia (1) o ausencia (0) de cada palabra, sin contar repeticiones.

**¿Cuándo usarlo?** Cuando la frecuencia de aparición no es relevante, solo si la palabra aparece o no. Útil para vocabularios pequeños o cuando queremos evitar que palabras muy repetidas dominen la representación.

**Diferencia con BoW:** Si 'película' aparece 3 veces en un documento, BoW pone 3, One-Hot pone 1.

```python
# binary=True activa One-Hot Encoding en CountVectorizer
ohe_vectorizer = CountVectorizer(binary=True)
ohe_matrix = ohe_vectorizer.fit_transform(corpus_prep)

ohe_df = pd.DataFrame(
    ohe_matrix.toarray(),
    columns=ohe_vectorizer.get_feature_names_out(),
    index=[f'Doc{i+1}' for i in range(len(corpus))]
)

print('Matriz One-Hot Encoding (1=presente, 0=ausente):')
print(ohe_df.to_string())
print()
print('Comparación BoW vs One-Hot (primeras 5 columnas):')
cols = list(vocabulario[:5])
print('BoW:     ', bow_df[cols].iloc[0].values)
print('One-Hot: ', ohe_df[cols].iloc[0].values)
```

## Representación 3 — TF (Term Frequency)

**¿Qué es?** La frecuencia de un término normalizada por la longitud del documento. Evita que documentos más largos tengan automáticamente valores más altos.

**Fórmula:** `TF(t, d) = (número de veces que t aparece en d) / (total de términos en d)`

**Diferencia con BoW:** BoW da conteos absolutos; TF da frecuencias relativas (proporciones entre 0 y 1).

```python
def calcular_tf(corpus_tokenizado):
    """
    Calcula TF para cada término en cada documento.
    TF(t,d) = count(t en d) / len(d)
    Retorna lista de diccionarios {término: tf_value}
    """
    tf_corpus = []
    for doc in corpus_tokenizado:
        tokens = doc.split()
        total = len(tokens)
        conteos = Counter(tokens)
        tf = {term: count / total for term, count in conteos.items()}
        tf_corpus.append(tf)
    return tf_corpus

tf_corpus = calcular_tf(corpus_prep)

print('TF por documento (top 3 términos por documento):')
for i, tf_doc in enumerate(tf_corpus, 1):
    top3 = sorted(tf_doc.items(), key=lambda x: x[1], reverse=True)[:3]
    print(f'Doc{i}: {top3}')

# Convertir a DataFrame para visualización
all_terms = sorted(set(t for doc in corpus_prep for t in doc.split()))
tf_matrix = [[tf_doc.get(t, 0) for t in all_terms] for tf_doc in tf_corpus]
tf_df = pd.DataFrame(tf_matrix, columns=all_terms,
                     index=[f'Doc{i+1}' for i in range(len(corpus))])
print()
print('Matriz TF (valores redondeados a 3 decimales):')
print(tf_df.round(3).to_string())
```

## Representación 4 — IDF (Inverse Document Frequency)

**¿Qué es?** Mide qué tan rara o informativa es una palabra en el corpus. Palabras que aparecen en muchos documentos tienen IDF bajo (son comunes, poco discriminativas). Palabras que aparecen en pocos documentos tienen IDF alto (son específicas, muy informativas).

**Fórmula:** `IDF(t) = log(N / df(t))` donde N = número de documentos, df(t) = documentos que contienen t.

**Intuición:** Si 'película' aparece en los 5 documentos, IDF = log(5/5) = 0. Si 'magistral' aparece solo en 1, IDF = log(5/1) = 1.6. Las palabras raras son más informativas.

```python
def calcular_idf(corpus_tokenizado):
    """
    Calcula IDF para cada término del vocabulario.
    IDF(t) = log(N / df(t))
    Usamos log natural. Añadimos +1 al denominador para evitar división por cero.
    """
    N = len(corpus_tokenizado)
    # Contar en cuántos documentos aparece cada término
    df = Counter()
    for doc in corpus_tokenizado:
        tokens_unicos = set(doc.split())
        df.update(tokens_unicos)

    idf = {term: math.log(N / (count + 1)) for term, count in df.items()}
    return idf

idf_valores = calcular_idf(corpus_prep)

# Ordenar por IDF para ver las más y menos informativas
idf_ordenado = sorted(idf_valores.items(), key=lambda x: x[1], reverse=True)

print('Términos con IDF más ALTO (más informativos/raros):')
for term, val in idf_ordenado[:8]:
    print(f'  {term:<20} IDF = {val:.4f}')

print()
print('Términos con IDF más BAJO (más comunes/menos informativos):')
for term, val in idf_ordenado[-5:]:
    print(f'  {term:<20} IDF = {val:.4f}')
```

## Representación 5 — TF-IDF

**¿Qué es?** Combina TF e IDF: pondera cada término por su frecuencia en el documento (TF) multiplicada por su rareza en el corpus (IDF). Resultado: términos frecuentes en un documento pero raros en el corpus reciben el mayor peso.

**Fórmula:** `TF-IDF(t, d) = TF(t, d) × IDF(t)`

**¿Por qué es mejor que BoW?** BoW da el mismo peso a 'película' (aparece en todos los docs) y 'magistral' (aparece en uno). TF-IDF penaliza 'película' y premia 'magistral', capturando mejor qué hace único a cada documento.

Usamos `TfidfVectorizer` de scikit-learn que implementa una versión optimizada con suavizado.

```python
# TfidfVectorizer: implementación optimizada de TF-IDF
# norm='l2': normaliza cada vector a longitud unitaria (buena práctica)
# smooth_idf=True: añade 1 al denominador del IDF para evitar división por cero
tfidf_vectorizer = TfidfVectorizer(norm='l2', smooth_idf=True)
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus_prep)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out(),
    index=[f'Doc{i+1}' for i in range(len(corpus))]
)

print('Matriz TF-IDF (valores redondeados a 4 decimales):')
print(tfidf_df.round(4).to_string())
print()

# Mostrar los términos más importantes por documento
print('Términos más importantes por documento (mayor TF-IDF):')
for i, row in tfidf_df.iterrows():
    top3 = row.nlargest(3)
    print(f'{i}: {list(zip(top3.index, top3.round(4).values))}')
```

## Representación 6 — N-gramas (Bigramas y Trigramas)

**¿Qué son?** Secuencias contiguas de N palabras. Un bigrama es una secuencia de 2 palabras, un trigrama de 3.

**¿Por qué son útiles?** Capturan contexto local que BoW ignora. 'no buena' y 'muy buena' tienen significados opuestos, pero BoW las trataría igual (mismas palabras, diferente orden). Los bigramas capturan esta diferencia.

**Ejemplo:** 'no recomendaría' es un bigrama que captura negación, algo que BoW pierde completamente.

```python
# ngram_range=(2,2): solo bigramas
# ngram_range=(1,2): unigramas + bigramas (más común en práctica)
# ngram_range=(2,3): bigramas + trigramas

# --- Bigramas ---
bigram_vec = CountVectorizer(ngram_range=(2, 2))
bigram_matrix = bigram_vec.fit_transform(corpus_prep)
bigramas = bigram_vec.get_feature_names_out()

print(f'Número de bigramas únicos: {len(bigramas)}')
print(f'Ejemplos de bigramas: {list(bigramas[:10])}')
print()

# --- Trigramas ---
trigram_vec = CountVectorizer(ngram_range=(3, 3))
trigram_matrix = trigram_vec.fit_transform(corpus_prep)
trigramas = trigram_vec.get_feature_names_out()

print(f'Número de trigramas únicos: {len(trigramas)}')
print(f'Ejemplos de trigramas: {list(trigramas[:10])}')
print()

# --- TF-IDF con bigramas (combinación poderosa) ---
tfidf_bigram_vec = TfidfVectorizer(ngram_range=(1, 2), max_features=20)
tfidf_bigram_matrix = tfidf_bigram_vec.fit_transform(corpus_prep)

print('Top 20 features TF-IDF con unigramas+bigramas:')
print(list(tfidf_bigram_vec.get_feature_names_out()))
```

## Similitud entre documentos con TF-IDF

Una aplicación directa de las representaciones vectoriales es calcular similitud entre documentos usando la **similitud coseno**. Como mencionó el profesor en clase, la similitud coseno es la métrica estándar para comparar vectores TF-IDF.

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculamos similitud coseno entre todos los pares de documentos
sim_matrix = cosine_similarity(tfidf_matrix)

sim_df = pd.DataFrame(
    sim_matrix.round(3),
    index=[f'Doc{i+1}' for i in range(len(corpus))],
    columns=[f'Doc{i+1}' for i in range(len(corpus))]
)

print('Matriz de similitud coseno (TF-IDF):')
print(sim_df.to_string())
print()
print('Interpretación:')
print('- Doc1 y Doc4 (reseñas positivas) deberían tener mayor similitud entre sí')
print('- Doc2 y Doc5 (reseñas negativas) deberían tener mayor similitud entre sí')
print('- Valores cercanos a 1 = muy similares, cercanos a 0 = muy distintos')
```

## Discusión de los resultados

### Comparación de métodos

| Método | Dimensión | Captura orden | Captura rareza | Interpretable | Uso recomendado |
|---|---|---|---|---|---|
| Bag of Words | V (vocab) | No | No | Sí | Baseline rápido |
| One-Hot | V (vocab) | No | No | Sí | Vocabularios pequeños |
| TF | V (vocab) | No | No | Sí | Documentos de longitud variable |
| IDF | V (vocab) | No | Sí | Sí | Análisis de rareza |
| TF-IDF | V (vocab) | No | Sí | Sí | Clasificación, búsqueda |
| N-gramas | V×N | Parcial | Opcional | Parcial | Análisis de frases, negación |

### Observaciones clave

**TF-IDF vs BoW:** La diferencia más notable fue que TF-IDF asignó peso 0 a términos que aparecen en todos los documentos (como 'película'), mientras que BoW les daba el mismo peso que a términos únicos. Esto hace que TF-IDF sea mucho más discriminativo para clasificación.

**N-gramas:** Con solo 5 documentos cortos, el número de trigramas únicos fue muy bajo (casi uno por documento), lo que ilustra el problema de la **maldición de la dimensionalidad**: con corpus pequeños, los n-gramas de orden alto son demasiado escasos para ser útiles estadísticamente.

**Similitud coseno:** Los documentos positivos (Doc1, Doc4) mostraron mayor similitud entre sí que con los negativos, validando que TF-IDF captura información semántica relevante incluso sin entender el significado de las palabras.

### Limitaciones

Todos estos métodos son **representaciones dispersas** (sparse): la mayoría de los valores son 0. Con vocabularios grandes (miles de palabras), esto es computacionalmente ineficiente. Los embeddings densos (Word2Vec, BERT) del Tema 5 resuelven este problema al representar palabras en espacios de menor dimensión pero más ricos semánticamente.

### Conclusión

Para la tarea de clasificación de reseñas, **TF-IDF con unigramas y bigramas** sería la representación más adecuada de las vistas en este tema: captura la rareza de los términos, normaliza por longitud del documento, y los bigramas permiten capturar negaciones ('no recomendaría'). Esta es la representación que usaremos en el Tema 4 con los modelos clásicos.

## Paso 3 — Entrenamiento y evaluación de los 4 modelos

Entrenamos cada modelo midiendo el tiempo de entrenamiento y predicción. Esto es importante para comparar no solo la calidad sino también la eficiencia computacional.

```python
def entrenar_evaluar(nombre, modelo, X_train, y_train, X_test, y_test):
    """
    Entrena un modelo, mide tiempos y calcula métricas de evaluación.
    Retorna diccionario con todos los resultados.
    """
    # Entrenamiento
    t0 = time.time()
    modelo.fit(X_train, y_train)
    t_train = time.time() - t0

    # Predicción
    t0 = time.time()
    y_pred = modelo.predict(X_test)
    t_pred = time.time() - t0

    # Métricas (macro: trata todas las clases por igual)
    return {
        'Modelo': nombre,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred, average='macro'),
        'Recall': recall_score(y_test, y_pred, average='macro'),
        'F1': f1_score(y_test, y_pred, average='macro'),
        'T_train(s)': round(t_train, 3),
        'T_pred(s)': round(t_pred, 3),
        'y_pred': y_pred
    }

# Definición de los 4 modelos
# Naive Bayes: requiere valores no negativos → compatible con TF-IDF
# Regresión Logística: max_iter=1000 para asegurar convergencia
# LinearSVC: versión eficiente de SVM para texto (más rápida que SVC)
# KNN: n_neighbors=5 estándar; metric='cosine' es mejor para texto
modelos = [
    ('Naive Bayes',         MultinomialNB(alpha=0.1)),
    ('Regresión Logística', LogisticRegression(max_iter=1000, C=1.0, random_state=42)),
    ('SVM (LinearSVC)',     LinearSVC(C=1.0, max_iter=2000, random_state=42)),
    ('KNN (k=5)',           KNeighborsClassifier(n_neighbors=5, metric='cosine', algorithm='brute'))
]

resultados = []
predicciones = {}

for nombre, modelo in modelos:
    print(f'Entrenando {nombre}...')
    res = entrenar_evaluar(nombre, modelo, X_train, y_train, X_test, y_test)
    predicciones[nombre] = res.pop('y_pred')  # guardar predicciones separado
    resultados.append(res)
    print(f'  ✓ F1={res["F1"]:.4f} | T_train={res["T_train(s)"]}s')

print('\n✓ Todos los modelos entrenados')
```

## Paso 4 — Tabla comparativa de métricas

En categorías temáticamente muy distintas con vocabularios separados, Naive Bayes puede competir o superar a modelos más complejos.

Para producción en un problema general de clasificación de texto, seguiría recomendando **Regresión Logística** como primera opción porque, además de rendimiento competitivo, produce probabilidades calibradas (no solo clases). ## 4. Preprocesamiento y corpus para embeddings

Definimos funciones de limpieza del texto y tokenización. Esto facilita entrenar embeddings y construir las secuencias para la RNN.

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt', quiet=True)

SPANISH_REMOVE = re.compile(r'[^a-zA-ZáéíóúÁÉÍÓÚüÜñÑ\s]')


def clean_text(text):
    text = text.lower()
    text = SPANISH_REMOVE.sub(' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text


def tokenize_text(text):
    return word_tokenize(text)

train_texts = [clean_text(doc) for doc in raw_train.data]
test_texts = [clean_text(doc) for doc in raw_test.data]

train_tokens = [tokenize_text(doc) for doc in train_texts]
test_tokens = [tokenize_text(doc) for doc in test_texts]

print('Ejemplo de texto limpio:')
print(train_texts[0][:300])
print('\nTokens:', train_tokens[0][:20])
```

## 5. Entrenamiento de Word2Vec

Entrenamos embeddings Word2Vec sobre el corpus tokenizado. Esto ilustra cómo se aprenden vectores de palabras a partir de contexto.

```python
EMBEDDING_SIZE = 100
WINDOW_SIZE = 5
MIN_COUNT = 2

word2vec_model = Word2Vec(
    sentences=train_tokens,
    vector_size=EMBEDDING_SIZE,
    window=WINDOW_SIZE,
    min_count=MIN_COUNT,
    workers=2,
    epochs=10,
    seed=RANDOM_SEED
)

print(f'Vocabulary size: {len(word2vec_model.wv)} palabras')
print('Vector de ejemplo para la palabra "space":')
print(word2vec_model.wv['space'][:10])
```

### 5.1. Exploración de embeddings

## 6. Definición del dataset para PyTorch

Construimos un dataset que convierte cada documento en una secuencia de índices y hace padding para la RNN.

```python
MAX_SEQ_LEN = 120

vocab = {word: idx + 1 for idx, word in enumerate(word2vec_model.wv.index_to_key)}
vocab['<PAD>'] = 0


def tokens_to_indices(tokens, vocab, max_len):
    indices = [vocab.get(token, 0) for token in tokens]
    if len(indices) > max_len:
        return indices[:max_len]
    return indices + [0] * (max_len - len(indices))

class TextDataset(Dataset):
    def __init__(self, token_lists, labels, vocab, max_len):
        self.token_lists = token_lists
        self.labels = labels
        self.vocab = vocab
        self.max_len = max_len

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        tokens = self.token_lists[idx]
        x = torch.tensor(tokens_to_indices(tokens, self.vocab, self.max_len), dtype=torch.long)
        y = torch.tensor(self.labels[idx], dtype=torch.long)
        return x, y

train_dataset = TextDataset(train_tokens, raw_train.target, vocab, MAX_SEQ_LEN)
test_dataset = TextDataset(test_tokens, raw_test.target, vocab, MAX_SEQ_LEN)

print('Ejemplo secuencia (primer documento):')
print(train_dataset[0][0][:20].tolist())
```

## 7. Modelo RNN con embeddings preentrenados

Creamos una RNN simple que utiliza la matriz de embeddings entrenada con Word2Vec. Esto muestra cómo se pueden aplicar embeddings a una tarea de clasificación.

```python
class RNNClassifier(nn.Module):
    def __init__(self, embedding_matrix, hidden_dim, output_dim, n_layers=1, dropout=0.3):
        super().__init__()
        num_embeddings, embedding_dim = embedding_matrix.shape
        self.embedding = nn.Embedding.from_pretrained(torch.tensor(embedding_matrix, dtype=torch.float32), freeze=False, padding_idx=0)
        self.rnn = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers, batch_first=True, dropout=dropout, bidirectional=True)
        self.classifier = nn.Sequential(
            nn.Dropout(dropout),
            nn.Linear(hidden_dim * 2, output_dim)
        )

    def forward(self, x):
        embedded = self.embedding(x)
        _, (hidden, _) = self.rnn(embedded)
        hidden_cat = torch.cat((hidden[-2], hidden[-1]), dim=1)
        return self.classifier(hidden_cat)

embedding_matrix = np.zeros((len(vocab), EMBEDDING_SIZE), dtype=np.float32)
for word, idx in vocab.items():
    if word == '<PAD>':
        continue
    if word in word2vec_model.wv:
        embedding_matrix[idx] = word2vec_model.wv[word]
    else:
        embedding_matrix[idx] = np.random.uniform(-0.1, 0.1, EMBEDDING_SIZE)

MODEL_HIDDEN_DIM = 64
OUTPUT_DIM = len(categorias)

model_rnn = RNNClassifier(embedding_matrix, MODEL_HIDDEN_DIM, OUTPUT_DIM)
print(model_rnn)
```

## 8. Entrenamiento del modelo RNN

Entrenamos el modelo en un conjunto pequeño para ilustrar el proceso. Se trata de una demostración reproducible y adecuada para un notebook educativo.

```python
BATCH_SIZE = 32
NUM_EPOCHS = 3
LEARNING_RATE = 1e-3

def train_model(model, train_dataset, val_dataset, epochs):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)
    optimizer = torch.optim.Adam(model.parameters(), lr=LEARNING_RATE)
    criterion = nn.CrossEntropyLoss()

    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE)

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0
        for x, y in train_loader:
            x, y = x.to(device), y.to(device)
            optimizer.zero_grad()
            logits = model(x)
            loss = criterion(logits, y)
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()
        print(f'Epoch {epoch+1}/{epochs} — loss={epoch_loss/len(train_loader):.4f}')

    model.eval()
    y_true, y_pred = [], []
    with torch.no_grad():
        for x, y in val_loader:
            x = x.to(device)
            logits = model(x)
            preds = logits.argmax(dim=1).cpu().numpy()
            y_true.extend(y.numpy())
            y_pred.extend(preds)

    return y_true, y_pred

# Usamos una muestra razonable para acelerar el entrenamiento
train_indices, _ = train_test_split(range(len(train_dataset)), train_size=400, random_state=RANDOM_SEED, stratify=raw_train.target)
val_indices, _ = train_test_split(range(len(test_dataset)), train_size=200, random_state=RANDOM_SEED, stratify=raw_test.target)

train_subset = torch.utils.data.Subset(train_dataset, train_indices)
val_subset = torch.utils.data.Subset(test_dataset, val_indices)

start_time = time.time()
y_true_rnn, y_pred_rnn = train_model(model_rnn, train_subset, val_subset, NUM_EPOCHS)
print(f'Tiempo de entrenamiento RNN: {time.time() - start_time:.1f}s')
```

```python
accuracy_rnn = accuracy_score(y_true_rnn, y_pred_rnn)
precision_rnn = precision_score(y_true_rnn, y_pred_rnn, average='weighted')
recall_rnn = recall_score(y_true_rnn, y_pred_rnn, average='weighted')
f1_rnn = f1_score(y_true_rnn, y_pred_rnn, average='weighted')

print('Resultados RNN con Word2Vec embeddings:')
print(f'Accuracy:  {accuracy_rnn:.4f}')
print(f'Precision: {precision_rnn:.4f}')
print(f'Recall:    {recall_rnn:.4f}')
print(f'F1-score:  {f1_rnn:.4f}')
```

## 9. Fine-tuning de un modelo Transformer

Aplicamos `DistilBERT` fine-tuneado para la misma tarea de clasificación. DistilBERT es un Transformer ligero que permite obtener buenos resultados con menor costo de cómputo.

```python
from datasets import Dataset

# Construimos un dataset de Hugging Face para entrenamiento y evaluación        
hf_train = Dataset.from_dict({
    'text': train_texts,
    'label': raw_train.target
})
hf_test = Dataset.from_dict({
    'text': test_texts,
    'label': raw_test.target
})

# Reducimos el tamaño del dataset para acelerar el notebook cuando se ejecute en CPU
hf_train = hf_train.shuffle(seed=RANDOM_SEED).select(range(400))
hf_test = hf_test.shuffle(seed=RANDOM_SEED).select(range(200))

model_name = 'distilbert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_fn(batch):
    return tokenizer(batch['text'], truncation=True, padding='max_length', max_length=128)

hf_train = hf_train.map(tokenize_fn, batched=True)
hf_test = hf_test.map(tokenize_fn, batched=True)

hf_train.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])
hf_test.set_format(type='torch', columns=['input_ids', 'attention_mask', 'label'])

model_transformer = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=len(categorias))

# Entrenamiento manual sin Trainer
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model_transformer.to(device)

train_loader = DataLoader(hf_train, batch_size=8, shuffle=True)
eval_loader = DataLoader(hf_test, batch_size=16)

optimizer = torch.optim.AdamW(model_transformer.parameters(), lr=2e-5)
criterion = nn.CrossEntropyLoss()

TRAIN_EPOCHS = 2

start_time = time.time()
for epoch in range(TRAIN_EPOCHS):
    model_transformer.train()
    total_loss = 0
    for batch in train_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)
        
        optimizer.zero_grad()
        outputs = model_transformer(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    print(f'Epoch {epoch+1}/{TRAIN_EPOCHS} - Loss: {total_loss/len(train_loader):.4f}')

train_duration = time.time() - start_time
print(f'Tiempo de entrenamiento Transformer: {train_duration:.1f}s')

# Evaluación
model_transformer.eval()
y_true_transformer, y_pred_transformer = [], []
with torch.no_grad():
    for batch in eval_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label']
        
        outputs = model_transformer(input_ids=input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = logits.argmax(dim=1).cpu().numpy()
        y_true_transformer.extend(labels.numpy())
        y_pred_transformer.extend(preds)

transformer_metrics = {
    'accuracy': accuracy_score(y_true_transformer, y_pred_transformer),
    'precision': precision_score(y_true_transformer, y_pred_transformer, average='weighted'),
    'recall': recall_score(y_true_transformer, y_pred_transformer, average='weighted'),
    'f1': f1_score(y_true_transformer, y_pred_transformer, average='weighted')
}
print('Transformer Metrics:', transformer_metrics)
```

## 2. Importaciones y datos de ejemplo

Creamos un corpus de oraciones en español con entidades de persona, organización, lugar y fecha.

```python
import re

from transformers import pipeline

texts = [
    'Madrid acogerá la próxima conferencia de IA organizada por IBM el 20 de mayo de 2026.',
    'La startup española Glovo anunció una colaboración con Telefónica en Barcelona.',
    'La película dirigida por Pedro Almodóvar ganó el premio en el Festival de Cannes.',
    'El presidente de México visitó la sede de la ONU en Nueva York la semana pasada.',
    'El Sevilla FC jugará contra el Real Madrid en el estadio Ramón Sánchez-Pizjuán.'
]

print('Corpus de ejemplos para NER:')
for i, texto in enumerate(texts, 1):
    print(f'[{i}] {texto}')
```

## 3. Enfoque clásico: reglas, diccionarios y expresiones regulares

Definimos un método basado en listas de entidades y patrones regulares para extraer fechas y lugares. Este enfoque es simple, interpretable y útil cuando se busca transparencia.

```python
PERSONAS = ['Pedro Almodóvar', 'El presidente de México']
ORGANIZACIONES = ['IBM', 'Glovo', 'Telefónica', 'ONU', 'Sevilla FC', 'Real Madrid', 'Festival de Cannes']
LUGARES = ['Madrid', 'Barcelona', 'Nueva York', 'México', 'Ramón Sánchez-Pizjuán']

DATE_PATTERN = re.compile(r'\b(\d{1,2} de [a-zA-Z]+ de \d{4}|la semana pasada|próxima conferencia)\b', re.IGNORECASE)


def extract_entities_classic(text):
    entities = []
    for persona in PERSONAS:
        if persona in text:
            entities.append({'entity': 'PER', 'value': persona, 'method': 'diccionario'})
    for org in ORGANIZACIONES:
        if org in text:
            entities.append({'entity': 'ORG', 'value': org, 'method': 'diccionario'})
    for lugar in LUGARES:
        if lugar in text:
            entities.append({'entity': 'LOC', 'value': lugar, 'method': 'diccionario'})
    for match in DATE_PATTERN.finditer(text):
        entities.append({'entity': 'DATE', 'value': match.group(0), 'method': 'regex'})
    return entities

for i, texto in enumerate(texts, 1):
    print(f'\n--- Texto {i} ---')
    for ent in extract_entities_classic(texto):
        print(ent)
```

### 3.1. Observaciones del enfoque clásico

- Es interpretable: cada extracción puede explicarse directamente.
- Funciona bien para entidades conocidas y patrones frecuentes.
- Es frágil ante variaciones de texto, errores tipográficos y nuevas entidades.
- No generaliza automáticamente a personas, organizaciones o lugares no listados.

## 4. Enfoque moderno: modelo Transformer para NER

Usamos un pipeline de Hugging Face con un modelo Transformer de NER en español.

```python
# dslim/bert-base-NER usa safetensors, compatible con torch 2.4
ner_pipeline = pipeline('ner', model='dslim/bert-base-NER', aggregation_strategy='simple')

for i, texto in enumerate(texts, 1):
    print(f'\n--- Texto {i} ---')
    ners = ner_pipeline(texto)
    for ent in ners:
        print(f"  {ent['entity_group']:6s} | {ent['word']:30s} | score: {ent['score']:.2f}")
```

## 5. Comparación de resultados

Comparamos el enfoque clásico y el Transformer sobre los mismos textos. Prestamos atención a:
- tipos de entidades detectadas
- exactitud de los límites
- errores de reconocimiento
- ausencia de entidades relevantes

```python
for i, texto in enumerate(texts, 1):
    classic = extract_entities_classic(texto)
    transformer = ner_pipeline(texto)
    print(f'\n--- Comparación texto {i} ---')
    print('  Clásico:')
    for e in classic:
        print(f"    {e['entity']:6s} | {e['value']}")
    print('  Transformer:')
    for e in transformer:
        print(f"    {e['entity_group']:6s} | {e['word']:30s} | score: {e['score']:.2f}")
```

## 6. Comparación de resultados y reflexión
- **Ventajas:** transparencia, bajo costo computacional, control absoluto sobre las entidades reconocidas.
- **Desventajas:** requiere mantenimiento manual, no adapta bien a textos nuevos ni a nombres no previstos.
- **Fallo típico:** si aparece una entidad con una variante no listada, no se detecta.

Ejemplo de fallo:
- si apareciera "Barcelona FC" en lugar de "Sevilla FC", el método clásico no lo identificaría sin actualizar el diccionario.

## 7. Discusión del enfoque Transformer

El modelo `dslim/bert-base-NER` fue entrenado en inglés (CoNLL-2003), lo que explica varios de los errores observados al aplicarlo a textos en español:

- **Fragmentación de tokens con acentos:** `Pedro Almodóvar` se detecta como `Pedro Almod` porque el tokenizador BERT no reconoce bien los caracteres acentuados del español, produciendo subwords como `##óvar` que no se agrupan correctamente.
- **Errores de etiquetado por idioma:** En el texto 4, `El` se clasifica como ORG y `##ueva` aparece como fragmento suelto. Esto ocurre porque el modelo no comprende la estructura sintáctica del español.
- **Confusión semántica:** `Ramón Sánchez-Pizjuán` (un estadio, LOC) se etiqueta como PER porque el modelo asocia ese patrón de nombre propio con personas en inglés.
- **Entidades parciales:** `Glovo` se detecta como `G` y `Barcelona` se clasifica como ORG en lugar de LOC, errores directamente relacionados con el desajuste lingüístico.

**Conclusión:** Un modelo NER entrenado en inglés aplicado a español produce resultados degradados. Para producción en español se debería usar un modelo específico como `PlanTL-GOB-ES/roberta-base-cap3-ner` o hacer fine-tuning sobre un corpus anotado en español.
